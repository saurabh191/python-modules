### Matplotlib Tutorial

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,30,5)
y = np.arange(0,18,3)
```

```python
##plotting using matplotlib

## plt scatter


plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
## plt plot

a = np.arange(0,30,5)
b=a*a
plt.plot(a,b,'r*--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('plot_test.png')
```

```python
# Creating Subplots
#plt.plot(totat_rows, total_cols, current_index)
plt.subplot(2,2,1)
plt.plot(a,b,'r--')
plt.subplot(2,2,2)
plt.plot(a,b,'b--')
plt.subplot(2,2,3)
plt.plot(a,b,'g-*')
plt.subplot(2,2,4)
plt.plot(a,b,'o-')
```

```python
y_sin=np.sin(np.arange(0,4*np.pi,0.1))
y_cos=np.cos(np.arange(0,4*np.pi,0.1))
plt.plot(y_sin, 'r--')
plt.plot(y_cos, 'g-*')
plt.title('sine-cos wave form')
```
